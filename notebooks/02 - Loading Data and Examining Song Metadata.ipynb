## Getting the Data

```python
import glob
import os
```

```python
file_path = os.path.join(os.path.pardir, "data", "MillionSongSubset", "data")
files = glob.glob(file_path + '/**/*.h5', recursive=True)
```

```python
files[:10]
```

```python
# Check the keys 
with h5py.File(files[0], "r") as f:
    print(list(f.keys()))
```

```python
# Examine each dataset
store = pd.HDFStore(files[0])
```
